# Data Acquisition

This notebook loads `creditcard.csv` into a pandas dataframe, converts each row to a MongoDB document, and inserts all 284,807 records into the `transactions` collection in MongoDB Atlas.

**Imports and logging:**

```python
import pandas as pd
from pymongo import MongoClient
import os
import logging

# Configure logging to record all steps to a log file
logging.basicConfig(
    filename='data_acquisition.log',
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
```

**Connect to MongoDB:**

```python
# Load credentials from environment variables and connect to MongoDB Atlas
try:
    user = os.environ.get("MONGOUSER")
    password = os.environ.get("MONGOPASS")
    client = MongoClient(f"mongodb+srv://{user}:{password}@cluster0.kcnfk.mongodb.net/?appName=Cluster0")
    logging.info("Connected to MongoDB Atlas successfully")
    print("Connected successfully")
except Exception as e:
    logging.error(f"Failed to connect to MongoDB: {e}")
    raise
```

**Load CSV:**

```python
# Path to creditcard.csv in OneDrive
CSV_PATH = '/Users/margauxreynolds/Library/CloudStorage/OneDrive-UniversityofVirginia/ds4320-project-2/creditcard.csv'

# Load creditcard.csv into a dataframe
try:
    df = pd.read_csv(CSV_PATH)
    logging.info(f"Loaded creditcard.csv with {len(df)} rows")
    print(f"Loaded {len(df)} rows")
except Exception as e:
    logging.error(f"Failed to load CSV: {e}")
    raise
```

**Insert into MongoDB:**

```python
# Convert dataframe rows to documents and insert into transactions collection
try:
    db = client["ds4320_project2"]
    collection = db["transactions"]
    collection.drop()
    logging.info("Dropped existing transactions collection")
    documents = df.to_dict(orient="records")
    collection.insert_many(documents)
    logging.info(f"Inserted {len(documents)} documents into transactions collection")
    print(f"Inserted {len(documents)} documents into transactions collection")
except Exception as e:
    logging.error(f"Failed to insert documents: {e}")
    raise
```

```python
# Verify documents were inserted correctly
try:
    count = collection.count_documents({})
    logging.info(f"Verified {count} documents in transactions collection")
    print(f"Verified: {count} documents in transactions collection")
except Exception as e:
    logging.error(f"Failed to verify insertion: {e}")
    raise
```
